```python
#依據etag種類分表格
NData['Detected'] = Detected
ADR = np.zeros([15,59])
for i in set(NData['EtagModel']):
    tempDF = NData.loc[np.where(NData['EtagModel'] == i)[0],['Detected','EtagModel','EtagExistDays']]
    tnum = 0
    for num in range(59):
        DT = tempDF.loc[tempDF['EtagExistDays'] < 30*(num+1),'Detected']
        if len(DT) == 0:
            ADR[int(i),num] = -1
            continue
        ADR[int(i),num] = sum(DT)/len(DT)
        tnum = tnum + len(DT)
        tempDF.loc[tempDF['EtagExistDays'] < 30*(num+1),'EtagExistDays'] = 2000
        if tnum == etagmodelnum[int(i)]:
            break
        
```

```python
# 把交易次數大於一百萬的etagmodel列出來
plotDR = ADR[np.asarray(etagmodelnum) > 1000000,:]
plotname = EtagModel_label.classes_[np.asarray(etagmodelnum) > 1000000]

# 依據30天唯一個間距畫圖

x = [(i+1)*30 for i in range(59)]

plt.figure(figsize=(30,7)) 
ax = plt.subplot(111)
# ax.xaxis.label.set_fontsize(15)
# ax.title.set_fontsize()
for i in range(7):
    plt.plot(x,plotDR[i,:],label = plotname[i],lw = 3)
    plt.ylim(0,1.1)
    plt.xlim(25,1800)
plt.xlabel('Days')
plt.ylabel('Detected Rate')
plt.title('etagmodel compare')
plt.legend(loc = 'best',fontsize = 10)
plt.xticks(x)

rc('xtick', labelsize=10) 


plt.show
    
```

```python
etagmodelnum = []
for i in set(NData['EtagModel']):
    etagmodelnum.append(len(np.where(NData['EtagModel'] == i)[0]))
    print(len(np.where(NData['EtagModel'] == i)[0]))

plt.figure(figsize=(13,7)) 
plt.bar(EtagModel_label.classes_, etagmodelnum)
# plt.xticks(indexes +  0.5, labels)
# plt.xticks(['010','040','110','111','112','113','114','115','116','120','120H','121','122','230','231'])
# ha: horizontal alignment
# va: vertical alignment
for x, y in zip(EtagModel_label.classes_, etagmodelnum):
    plt.text(x , y ,'{}'.format(y), ha='center', va='bottom', fontsize=12)
plt.show()
```

# 使用gridsearch找最佳參數
## 結果是'min_impurity_decrease': 1e-05,'min_samples_leaf': 1000是最好的，並且沒有overfitting的問題，時間上的差異主要是n_estimators的大小，越大越好(差異不太大)，但時間成本也高，考慮效益，改用50就好

實際上我使用fully-growth tree，因先不看overfitting的問題，只找影響較大的因子
